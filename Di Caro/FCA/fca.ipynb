# Formal Component Analysis

```python
from concepts import Context
import csv

path = 'resources/concepts.csv'
with open(path, 'r', encoding='utf8') as f:
    data = csv.reader(f)
    head = next(data)
    c = ",".join(head) + '\n'
    for i, row in enumerate(data):
        c += ",".join(row) + '\n'

r = Context.fromstring(c,frmat='csv')
print(r)
```
